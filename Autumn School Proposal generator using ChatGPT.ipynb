# NanoDTC Autumn School Mini Proposal generator using Chat GPT

What this script does:

1. Reads the csv file containing information submitted in the google doc form prior to the workshop. e.g:

| Student     | Email | Department | Supervisor | Project Title | Project | Project KeyWords |
| ----------- | ----------- | ----------- | ----------- | ----------- | ----------- | ----------- |
| Timothy Lambden     | tpgl3@cam.ac.uk | Material Sciences   | Paul Midgley | CryoSTEM | Project on CryoSTEM | Electron Microscopy, TEM, SEM, CryoEM 
| Lorenzo Peri   | lp586@cam.ac.uk | Physics/Quantum Motion | Fernando | Quantum computing with silicon | Qunatum computing using silicon qunatum dots and machine learning | quantum dots |  


2. Spin the Project randomiser! Selecting two students at random and outputs their information into a mini project proposal form, a potential pairing for a mini project!

3. Inputs this data into a a word document mini proposal form.

4. Information of the students and their research is then fed into ChatGPT and used to generate a project, namely a title and description of project, e.g. abstract, obejctives, and conclusion (this requires an openAI API key, which currently you have to pay for).

4. An image is generated based using Dalle based on the project title. an AI text-to-image generator and asks for an image (often this doesn't work too well).

5. This information is put into a document for a potential mini project... It would likely be a bad project, but may spark an idea for a potential collaboration of a project.

What this script doesn't do:

1. Produce a cohert, readable and well thought mini project proposal.
2. Explain how it generated it. ChatGPT is quite a black box.
